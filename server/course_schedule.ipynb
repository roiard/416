```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from random import randint
import os
```

# First, randomly distribute the courses

```python
df = pd.read_excel('./course_22_F_CSE.xlsx')

df.head()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Total time'] = ((pd.to_datetime(df['End Time'].astype('string')) - pd.to_datetime(df['Start Time'].astype('string')))).dt.total_seconds() / 60
```

```python
df['Start Time'], df['End Time'] = None, None

df_filtered = df[df['Total time'] >0]
```

```python
start_time = {
    0: '9:00 AM',
    1:'10:30 AM', 
    2: '12:30 PM',
    3: '2:00 PM',
    4: '3:30 PM',
    5: '5:00 PM'
    }

num = df_filtered.shape[0] // 6
extra = df_filtered.shape[0] % 6

lst = [num for _ in range(6)]

for _ in range(extra):
    idx = randint(0, num-1)
    lst[idx] += 1

time_slots = []

for i in range(6):
    time_slots += [start_time[i]] * lst[i]
             
np.random.shuffle(time_slots)         
df_filtered['Start Time'] = time_slots
```

```python
def calculate_end_time(row):

    start_time_dt = datetime.strptime(row['Start Time'], '%I:%M %p')
    end_time_dt = start_time_dt + timedelta(minutes=row['Total time'])
    end_time_str = end_time_dt.strftime('%I:%M %p')

    return end_time_str

df_filtered['End Time'] = df_filtered.apply(calculate_end_time, axis=1)

df_filtered
```
